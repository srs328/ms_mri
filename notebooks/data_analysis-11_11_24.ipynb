```python
import pandas as pd
from pathlib import Path
import os
```

```python
secure_data_dir = Path("/mnt/h/Secure_Data")
df = pd.read_stata(secure_data_dir / "ACTRIMS_abstract_data.dta")
```

```python
df.head()
```

```python
import statsmodels.api as sm

df['sex_factor'], _ = df['var18'].factorize()
df['phenotype_factor'], _ = df['var19'].factorize()
predictors = ["age", "ICV", "sex_factor", "phenotype_factor"]
outcome = "pineal_vol"

x = df[predictors]
Y = df[outcome]

mod = sm.OLS(Y, x, missing='drop')
res_prob = mod.fit()
res_prob.summary()
```
