```python
import statsmodels.api as sm

df['sex_factor'], _ = df['var18'].factorize()
df['phenotype_factor'], _ = df['var19'].factorize()
predictors = ["age", "ICV", "sex_factor", "phenotype_factor"]
outcome = "pituitary_vol"

x = df[predictors]
Y = df[outcome]

mod = sm.OLS(Y, x, missing='drop')
res_prob = mod.fit()
res_prob.summary()
```
