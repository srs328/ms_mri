```python
import pandas as pd
from pathlib import Path
import json
```

```python
resource_dir = Path("/home/srs-9/Projects/ms_mri/LONG_THOMAS/resources")
```

```python
with open(resource_dir / "all_longitudinal_sessions.json", 'r') as f:
    subject_sessions = json.load(f)

session_list = []
for subid, sessions in subject_sessions.items():
    session_list.extend([f"sub{subid}/{ses}" for ses in sessions])

with open(resource_dir / "subjects_allses_long.txt", 'w') as f:
    f.writelines("\n".join(session_list))
```

```python
subject_sessions = pd.read_csv(resource_dir/"longitudinal_sessions.csv", index_col="subid", dtype={'ses1': 'Int64', 'ses2': 'Int64'})

session_list = []
for subid, row in subject_sessions.iterrows():
    session_list.extend([f"sub{subid}/{ses}" for ses in [row['ses1'], row['ses2']]])

with open(resource_dir / "subjects_sesssions_long.txt", 'w') as f:
    f.writelines("\n".join(session_list))
```
